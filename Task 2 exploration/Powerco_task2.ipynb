### Data visualization


```python
gas_df = client_df[['id', 'has_gas', 'churn']].groupby(['churn','has_gas'])['id'].count().unstack(level=0)
pgas_df = (gas_df.div(gas_df.sum(axis=1), axis=0))*100
pgas_df
```

```python
plots = pgas_df.plot(kind='bar', stacked=True, figsize=(8,8), rot=0)
pad = 1
for p in plots.patches:
        
        # Calculate annotation
        value = str(round(p.get_height(),1))
        # If value is 0 do not annotate
        if value == '0.0':
            continue
        plots.annotate(value,
                    ((p.get_x()+ p.get_width()/2)*pad-0.05, 
                     (p.get_y()+p.get_height()/2)*pad),
                    color='black',
                    size=15
                   )
```

```python
origin_df = client_df[['id', 'origin_up', 'churn']].groupby(['churn','origin_up'])['id'].count().unstack(level=0)
porigin_df = (origin_df.div(origin_df.sum(axis=1), axis=0))*100
porigin_df
```

```python
plots = porigin_df.plot(kind='bar', stacked=True, figsize=(8,8), rot=90)
pad = 1
for p in plots.patches:
        
        # Calculate annotation
        value = str(round(p.get_height(),1))
        # If value is 0 do not annotate
        if value == '0.0':
            continue
        plots.annotate(value,
                    ((p.get_x()+ p.get_width()/2)*pad-0.05, 
                     (p.get_y()+p.get_height()/2)*pad),
                    color='black',
                    size=15
                   )
```

```python
channel_df = client_df[['id', 'channel_sales', 'churn']].groupby(['churn','channel_sales'])['id'].count().unstack(level=0)
channel_df = (channel_df.div(channel_df.sum(axis=1), axis=0))*100
channel_df
```

```python
plots = channel_df.plot(kind='bar', stacked=True, figsize=(8,8), rot=90)
pad = 1
for p in plots.patches:
        
        # Calculate annotation
        value = str(round(p.get_height(),1))
        # If value is 0 do not annotate
        if value == '0.0':
            continue
        plots.annotate(value,
                    ((p.get_x()+ p.get_width()/2)*pad-0.05, 
                     (p.get_y()+p.get_height()/2)*pad),
                    color='black',
                    size=15
                   )
```

```python
client_df2 = client_df.groupby('churn').sum().transpose().reset_index()
client_df2.columns = ['var', 'retension', 'churn']
client_df2 = client_df2.loc[[0, 1, 2, 10 ]]
client_df2['total']=client_df2[['retension','churn']].sum(axis=1)
client_df2['ret'] = client_df2['retension']/client_df2['total']*100
client_df2['chu'] = client_df2['churn']/client_df2['total']*100
client_df2
```

```python
df_plot = client_df2[['var', 'ret', 'chu']]
df_plot
```

```python
contract_type = client_df[['id', 'has_gas', 'churn']]
contract = contract_type.groupby([contract_type['churn'], contract_type['has_gas']])['id'].count().unstack(level=0)
contract_percentage = (contract.div(contract.sum(axis=1), axis=0) * 100).sort_values(by=[1], ascending=False)
contract_percentage
```

```python
ax = df_plot.plot.barh(x='var', stacked=True, figsize=(14,8))
pad = 1
for p in ax.patches:
        
        # Calculate annotation
        value = str(round(p.get_width(),1))
        # If value is 0 do not annotate
        if value == '0.0':
            continue
        ax.annotate(value,
                    ((p.get_x()+ p.get_width()/2)*pad-0.05, 
                     (p.get_y()+p.get_height()/2)*pad),
                    color='black',
                    size=15
                   )
       
```
